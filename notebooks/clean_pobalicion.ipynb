```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os; import json
from simpledbf import Dbf5

dir = '/home/mitch/school/data/mexico_enigh/'
data_dir = '/home/mitch/Dropbox/data/mexico_enigh/raw/1994/'
out_dir = '/home/mitch/Dropbox/data/mexico_enigh/interim/1994/'
```

```python
os.chdir(data_dir)
pobalicion = Dbf5('POBLA94.dbf').to_dataframe()

os.chdir(dir + 'notebooks/dicts/')
with open('pobalicion.json', 'r') as fp:
    names = json.load(fp)
os.chdir(out_dir)
```

```python
pobalicion = pobalicion.rename(columns=names)
toint = ['relationship_to_hhm', 'sex', 'age', 'type_of_education', 'education', 'education_technical']
pobalicion[toint] = pobalicion[toint].astype('int')

pobalicion = pobalicion.query('relationship_to_hhm == 1')
pobalicion
```

```python
os.chdir(out_dir)
keep = ['hhid', 'sex', 'age', 'education']
pobalicion[keep].to_csv('population_cleaned.csv')
```
