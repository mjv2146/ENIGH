```python
import hh_size
concen['hh_size_kids'] = concen['younger_members']
concen['hh_size_adults'] = concen['household_members'] - concen['hh_size_kids']
concen['hh_size'] = concen['total_population']
concen['hh_size_oecd'] = concen.apply(lambda x : hh_size.oecd_size(x.hh_size_adults, x.hh_size_kids), axis=1)
keep_concen = keep_concen + ['hh_size_oecd']
```
